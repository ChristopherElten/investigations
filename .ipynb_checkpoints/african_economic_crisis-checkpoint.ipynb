First group by subject descriptor and units b/c some descriptors have two units they are described with
eg. Current account balance.  For each group create an object that has what you need for a plot: axis labels, units, x-axis values, and the y-axis values that I call 'data'.  Note that I've made x_axis and data lists.  There's a reason for that ie. lists can be made in to whatever you want, easier to work with in pure python.

```python
def pluck(x):
    label = x.Country.to_list()
    y_label = x.Units.iloc[0]
    x_label = "Date"
    dates = ['1980', '1981', '1982', '1983', '1984',
       '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993',
       '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002',
       '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011',
       '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020',
       '2021', '2022', '2023', '2024']
    x_axis = [pd.to_datetime(date).year for date in dates]
    data = {}
    for country in x.Country:
        data[country] = x.loc[x.Country==country, dates].iloc[0].tolist()
    z = {y_label: {"label": label, "x_label": x_label, "y_label": y_label, "x_axis": x_axis, "data": data}}
    return z
    

packets = x.groupby(['Subject Descriptor', 'Units']).apply(lambda x: pluck(x))
```

You can see the structure at different levels

```python
print(packets.keys(),'\n',packets[1].keys(),'\n',packets[1]['U.S. dollars'].keys())
```

Substitute the description and units as done below and the plot happens.  Not most elegant, probably better ways to do it but I had objects on my mind so ...

```python
y = packets['Inflation, average consumer prices']['Index']['Index']
# y = packets['Inflation, average consumer prices']["Percent change"]["Percent change"]

obj = y

a4_dims = (16, 12)
fig, ax = pyplot.subplots(figsize=a4_dims)
pyplot.setp(ax.get_xticklabels(), rotation=45)

for country in obj['label']:
    data = pd.DataFrame([obj["x_axis"], obj['data'][country]]).T
    data.columns =[obj["x_label"], obj["y_label"]]
    sns.lineplot(data=data, x=obj["x_label"], y=obj["y_label"], legend='full')

pyplot.legend(labels=[country for country in obj['label']])
```
